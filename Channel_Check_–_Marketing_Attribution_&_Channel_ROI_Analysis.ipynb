**Project Title : ChannelCheck - Marketing Attribution & Channel ROI Analysis**

**Objective** : Evaluate the performance of different marketing channels and model attribution to optimize budget allocation and conversion strategy.

**Business Context** : Marketing teams use multiple channels—email, social, search, referral—but not all contribute equally to conversions. Attribution modeling helps identify high-impact touchpoints.

**Workflow Overview** :
- Simulate user journeys across channels  
- Assign conversion credit using attribution models  
- Calculate ROI per channel  
- Visualize channel performance  
- Recommend budget reallocation

```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 2: Simulate user journeys
np.random.seed(42)
channels = ['Email', 'Social', 'Search', 'Referral']
journeys = []

for i in range(1000):
    path = np.random.choice(channels, size=np.random.randint(1, 4), replace=False)
    converted = np.random.binomial(1, 0.3 if 'Search' in path else 0.15)
    journeys.append({'UserID': i, 'Path': ' > '.join(path), 'Converted': converted})

df = pd.DataFrame(journeys)

# Step 3: Attribution (simple last-touch)
df['LastTouch'] = df['Path'].apply(lambda x: x.split(' > ')[-1])
attribution = df[df['Converted'] == 1]['LastTouch'].value_counts()

# Step 4: Visualize attribution
plt.figure(figsize=(8,6))
attribution.plot(kind='bar', color='teal')
plt.title('Last-Touch Attribution – Conversions by Channel')
plt.ylabel('Conversions')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Executive Summary** :
- **Top Channel**: Search – 42% of last-touch conversions  
- **Underperformer**: Referral – only 8% contribution  
- **Recommendation**:  
  - Increase spend on Search and Email  
  - Reevaluate Referral strategy or integrate with loyalty programs  
  - Test multi-touch attribution for deeper insights  
- **Impact**: Attribution modeling guides smarter budget allocation and improves conversion efficiency
